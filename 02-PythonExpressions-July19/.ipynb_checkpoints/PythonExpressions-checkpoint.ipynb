#### Question 1
Write a function named `calculate_interest` that takes the following three arguments:

* `principal` (amount of money deposited or borrowed),
* `rate` (rate of interest per year, as a percentage),
* `time` (duration in years),

and returns the simple interest

#### **Example Usage:**

python
calculate_interest(1000, 5, 2)   # Output: 100.0 <br>
calculate_interest(1500, 4.3, 3) # Output: 193.5 <br>
calculate_interest(500, 10, 0)   # Output: 0.0 <br>

````python
def calculate_interest(principal,rate,time):
    return (principal*rate*time)/100
````

````python
principal = float(input("Enter the principal amount: "))
rate = float(input("Enter the interest rate (in %): "))
time = float(input("Enter the time (in years): "))
SimpleInterest=calculate_interest(principal,rate,time)
print(SimpleInterest)
````

#### 🔧 Task:

Write a function named `calculate_hypotenuse` that takes two arguments:

* `base`: the length of the base (a number)
* `height`: the length of the height/perpendicular side (a number)

It should return the length of the hypotenuse (a number).

#### 💡 Examples:

```python
Input: calculate_hypotenuse(3, 4)
Output: 5.0
```

Why?
Because √(3² + 4²) = √(9 + 16) = √25 = 5.0

---

```python
Input: calculate_hypotenuse(5, 12)
Output: 13.0
```

Why?
Because √(25 + 144) = √169 = 13.0

---

```python
Input: calculate_hypotenuse(0, 0)
Output: 0.0
```

#### PROOF


consider the  ΔABC and ΔADB,<br>
In  ΔABC and ΔADB,<br>
∠ABC = ∠ADB = 90°<br>
  ∠A = ∠A → common<br>
Using the AA criterion for the similarity of triangles, <br>
ΔABC ~ ΔADB <br>
Therefore, AD/AB = AB/AC<br>
⇒ AB2 = AC x AD  ……(1)<br>
<br>

∠C = ∠C  → common<br>
 ∠CDB = ∠ABC = 90°<br>
Using the Angle Angle(AA) criterion for the similarity of triangles, we conclude that,<br>
ΔBDC ~ ΔABC<br>
Therefore, CD/BC = BC/AC<br>
⇒ BC2 = AC x CD …..(2)<br>
From the similarity of triangles, we conclude that,<br>
∠ADB  = ∠CDB = 90°<br>
So if a perpendicular is drawn from the right-angled vertex of a right triangle to the hypotenuse, <br>
then the triangles formed on both sides of the perpendicular are similar to each other and also to the whole triangle.<br>
To Prove: AC2 =AB2 +BC2<br>
By adding equation (1) and equation (2), we get:<br>
AB2 + BC2= (AC x AD) + (AC x CD)<br>
AB2 + BC2 =  AC (AD + CD) …..(3)<br>
Since AD + CD = AC, substitute this value in equation (3).<br>
AB2 + BC2=  AC (AC)<br>
Now, it becomes<br>
AB2+ BC2=  AC2<br>
Hence, Pythagoras theorem is proved.<br>

````python
import math
def calculate_hypotenuse(base,height):
    return math.sqrt(base**2 + height**2)
    
````

````python
base = float(input("The length of the base (a number) "))
height = float(input("The length of the height/perpendicular side (a number)"))
Hypotenuse=calculate_hypotenuse(base,height)
print(Hypotenuse)
````

#### Find Distance 3D
Topic: Find Distance in 3D<br>
<br>
🧠 Explanation:<br>
In everyday life, we often measure the distance between two points—like how far your house is from school. In 3D space, 
we do the same, but we also consider height or depth in addition to length and width.<br>

A point in 3D space is defined using three coordinates:<br>
<br>
x (horizontal)<br>
y (vertical)<br>

To calculate the distance between two points in 3D, say:<br>
<br>

Come up with 3D distance formula using the Pythagorean theorem into three dimensions.<br>
<br>

Write a function called calculate_distance_3d(x1, y1, z1, x2, y2, z2) that returns the distance between two 3D points.<br>

You can use the built-in math.sqrt() function to calculate the square root.<br>
✅ Examples:<br>
<br>

#### Proof
We know distance between 2 point, now extending that to 3D<br>
![My Image](../Static/3D.png)
<br>
to reach the point Z, we will have to travel x then y, then z <br>
first using x and y calculate the hypotenuse of it, now this hypotenuse becomes another distance in<br>
the new triangle formed of sides new hypotenuse, z and the distance we wish to find.<br>
calculate the hypotenuse distance again and this time you will get the distance of the point in 3D

````python
def calculate_distance_3d(x1, y1, z1, x2, y2, z2):
    Zdistance=max(z1,z2)-min(z1,z2)
    NewDistance=find_distance_2d(x1,y1,x2,y2)
    return calculate_hypotenuse(Zdistance,NewDistance)
````

#### Find Manhattan Distance
Topic: Find Manhattan Distance<br>
Explanation:<br>
When you want to measure the distance between two points on a grid (like in a city with square blocks), you don’t move<br> diagonally. Instead, you move only horizontally and vertically, like a taxi driving along streets and avenues. This kind of<br> distance is called the Manhattan Distance.<br>
Manhattan Distance=|x1−x2|+|y1−y2|<br>
Exercise:<br>
Write a function manhattan_distance(x1, y1, x2, y2) that takes the coordinates of two points and returns their Manhattan<br> Distance.<br>
<br>
Example:<br>
manhattan_distance(2, 3, 5, 1)  # Output: 5<br>
manhattan_distance(0, 0, 0, 0)  # Output: 0<br>
manhattan_distance(-1, -1, 1, 1)  # Output: 4<br>

#### Proof
![My Image](../Static/Manhattan.png)
basically in this, we need to calculate x & y, that's why we do x1-x2 and y1-y2 and get abs of that.<br>
In previous methods, after doing this we used to find the third side, which is not required here. <br>
The blue and green line distance are always same, given we are trying to get min distance in Manhattan

````python
def manhattan_distance(x1, y1, x2, y2):
    X=abs(x1-x2)
    Y=abs(y1-y2)
    return X+Y
````

````python
manhattan_distance(2, 3, 5, 1) # Output: 5
manhattan_distance(0, 0, 0, 0) # Output: 0
manhattan_distance(-1, -1, 1, 1) # Output: 4
````

````python
def manhattan_distance(x1, y1, x2, y2):
    X=max(x1,x2)-min(x1,x2)
    Y=max(y1,y2)-min(y1,y2)
    return X+Y
````
